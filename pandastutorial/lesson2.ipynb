```python
import matplotlib.pyplot as plt
x = [5, 24, 30]
y = [5,6,8]

a = [2,3,4,5,6]
b = [3,6,7,9,10]

# plt.plot(x,y, marker = "o", linestyle='dotted')
# x, o, s, *, + , P, d, p, v, <,>, 1,2,3,4, | , _ , H
plt.plot(x,y, marker = "_", linestyle='dotted', color='red')
plt.plot(a,b, color ='green', marker='h', linestyle='dashed')
plt.xlabel('Number of example')
plt.ylabel('Sum of Data', fontdict={'color':'blue', 'size':14, 'family':'serif'})
plt.title('Polt tutorial', fontdict={'color':'red', 'size':14, 'family':'serif'}, loc='left')
# plt.grid()
# plt.grid(axis='x')
plt.grid(axis='y', color='blue', linestyle ='--', linewidth=0.25)
```

```python
import matplotlib.pyplot as plt
x = [5, 24, 30]
y = [5,6,8]

a = [2,3,4,5,6]
b = [3,6,7,9,10]

# plt.subplot(row, col, index) row = 2, col=3
plt.subplot(2,3,1)
plt.plot(x,y)

plt.subplot(2,3,2)
plt.plot(a,b)
plt.title('this is 2')
plt.xlabel('num of data')

plt.subplot(2,3,3)
plt.plot(a,b)

plt.subplot(2,3,4)
plt.plot(x,y)

plt.subplot(2,3,5)
plt.plot(a,b)

plt.subplot(2,3,6)
plt.plot(x,y)
plt.title('number 6')

plt.suptitle('this all plot title')
```
