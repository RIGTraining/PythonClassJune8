```python
import sqlite3
from tkinter import *
import tkinter as tk
import ttkbootstrap as ttk
from tkinter import messagebox
from ttkbootstrap.tableview import Tableview
from ttkbootstrap.constants import *

root = tk.Tk()
root.title('Contact App')
root.geometry('500x500')

conn = sqlite3.connect('mycontact.db')
c = conn.cursor()
c.execute('''
          CREATE TABLE IF NOT EXISTS contact(
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            name TEXT,
            phone TEXT,
            address TEXT)
          ''')
conn.commit()
conn.close()

def view_contact():
  conn = sqlite3.connect('mycontact.db')
  c = conn.cursor()
  data = c.execute("SELECT * FROM contact")
  rows = data.fetchall()  # fetch all rows from the executed query
  
  
  for row in rows:
    list.insert(END, f"{row[0]},{row[1]},{row[2]},{row[3]}")
    dt.insert_row('end', [row[1], row[2], row[3]])
  


def delete_contact():
  select_name = list.get(tk.ACTIVE)
  get_id = select_name.split(",")[0]
  print(get_id)
  conn = sqlite3.connect('mycontact.db')
  c = conn.cursor()
  c.execute("DELETE FROM contact WHERE id =?",(get_id,))
  conn.commit()#save changes
  conn.close()
  messagebox.showerror('success', 'Contact Delete')
  list.delete(0, END)
  view_contact()

def update_contact():
  select_name = list.get(tk.ACTIVE)
  get_id = select_name.split(",")[0]
  name = ename.get()
  ph = eph.get()
  ad = eadd.get()
  conn = sqlite3.connect('mycontact.db')
  c = conn.cursor()
  c.execute("update contact set name=?, phone=?, address=? where id=?",  (name,ph,ad, get_id))
  conn.commit()#save changes
  conn.close()
  messagebox.showinfo('success', 'Update Success')
  list.delete(0, END)
  ename.delete(0, END)
  eph.delete(0, END)
  eadd.delete(0, END)
  view_contact()



def save_data():
    name = ename.get()
    ph = eph.get()
    ad = eadd.get() 
    
    conn = sqlite3.connect('mycontact.db')
    c = conn.cursor()
    c.execute("INSERT INTO contact(name, phone, address) VALUES(?,?,?)",(name,ph,ad))
    conn.commit()#save changes
    conn.close()
    messagebox.showinfo('success', 'Contact Create Success')
    
    
lname= ttk.Label(text='Name :')
lph= ttk.Label(text='Phone :')
ladd= ttk.Label(text='Address :')
list = tk.Listbox(width=400)

ename = ttk.Entry(bootstyle="success")
eph = ttk.Entry(bootstyle="success")
eadd = ttk.Entry(bootstyle="success")

create_btn = ttk.Button(text='Create',bootstyle='success', command=lambda:save_data())
update_btn = ttk.Button(text='Update',bootstyle='info',command=lambda:update_contact())
del_btn = ttk.Button(text='Delete',bootstyle='danger', command=lambda:delete_contact())

lname.pack()
ename.pack()
lph.pack()
eph.pack()
ladd.pack()
eadd.pack()
create_btn.pack()
update_btn.pack()
del_btn.pack()
list.pack()


# colors = root.style.colors
coldata = [
    {"text": "Name", "stretch": False},
    "Phone",
    {"text": "Address", "stretch": False},
]

rowdata = []

dt = Tableview(
    master=root,
    coldata=coldata,
    rowdata=rowdata,
    paginated=True,
    searchable=True,
    bootstyle=PRIMARY,
    
)
dt.pack(fill=BOTH, expand=YES, padx=10, pady=10)







view_contact()
root.mainloop()
```
